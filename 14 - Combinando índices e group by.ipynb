# Utilizando o Apply para aplicar funções em Dataframes

```python
def consumo(a):
    if a <= 75:
        return 'LOW'
    if a >= 151:
        return 'HIGH'
    else:
        return 'MIDDLE'
```

```python
drinks['consumo'] = drinks['beer_servings'].apply(consumo)
```

# Para usar outro indice depois de ter agrupado, só criando um novo.

```python
drinks_4 = drinks.set_index(['consumo','continent'])
```

```python
drinks_4.head()
```

```python
# Aplicando Group By pelos dois indices 

drinks_4 = drinks_4.groupby(level = ['consumo','continent'])
```

```python
drinks_4.sum()
```

# Realizando a mesma operação com Numpy

```python
# Usando 'aggregate' do Pandas e 'np.sum' do NumPy

drinks_4.aggregate(np.sum)
```

```python
# O 'aggregate' pode agregar por um monte de medidas ao mesmo tempo
# (1) consumo = nível de consumo
# (2) continent = O continente em questão
# (3) sum = quantidade total de litros no continente
# (4) mean = quantidade média de litros no continente
# (5) size = total de paizes 

drinks_4.aggregate([np.sum, np.mean, np.size])
```
