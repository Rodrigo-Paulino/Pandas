# Ordenando Dataframes

```python
import numpy as np
import pandas as pd
```

```python
colunas_filmes = ['star_rating','title','content_rating','genre','duration','actors_list']

file="C:\\. ProjetosDataScience\\Datasets\\movies.csv"
filmes = pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns 
```

```python
filmes.head()
```

```python
#Ordenando por Título

filmes.title.sort_values().head(10)
```

```python
#O retorno é uma série

A = filmes.title.sort_values() # O retorno do Sort_Values é por padrão uma Série = 1 coluna com valores 
type(A)
```

```python
A
```

```python
#Ordem descendente

filmes.title.sort_values(ascending=False).head(10)
```

```python
#Há outra forma de ordenar um DataFrame. Desta forma, o retorno é um DataFrame ordenado por Título (Title).

filmes.sort_values('title').head(50)
```

```python
#Podemos também ordenar de forma descendente

filmes.sort_values('title', ascending=False).head(20)
```

```python
#Apenas ordenar não altera o DataFrame original

filmes.head()
```

```python
#Ordenando por classificação e duração (Primeiro ordena por Gênero depois por Duração)

filmes.sort_values(['genre', 'duration']).head(40)
```

## Filtrando um DataFrame

```python
#Vamos criar uma lista para saber os filmes com maior duração do que 180 minutos

boleanos = []

for item in filmes.duration:
    if item >= 200:
        boleanos.append(True)
    else:
        boleanos.append(False)
```

```python
boleanos[:20]
```

```python
type(boleanos)
```

```python
#Verificando o tamanho da lista
len(boleanos)
```

```python
filmes.shape
```

Precisamos converter a lista em uma SERIE para aplicar como filtro de Dataframe Pandas

```python
#Convertendo a lista e serie

duracao = pd.Series(boleanos)
duracao.head()
```

```python
type(duracao)
```

```python
#Podemos usar a notação de colchetes para indexar ou para aplicar filtros de series boleanas

print(filmes['title'][:10])
```

```python
filmes[duracao]
```

```python
#Podemos ainda continuar a notação trazendo somente as colunas desejadas

filmes[duracao][['title','star_rating']]
```

Podemos simplificar o filtro conforme abaixo ou ainda escrevendo em List Comprehension

```python
#Forma simplificada

A = filmes.duration >= 180
A.head(20)
```

```python
filmes[A].head()
```

```python
#Escrevendo por List Comprehension
# [] define lista, então jogar uma expressão dentro de umas lista (List Comprehension)

filmes[filmes.duration >= 180].head()
```

```python
#Podemos também retornar outra série com a coluna que desejarmos

genero_maior_200 = filmes[filmes.duration >= 200].genre #Usando Dot.Notetion
```

```python
genero_maior_200
```

```python
type(genero_maior_200)
```

```python
#Podemos ainda utilizar a notação de colchetes para fazer a mesma coisa

genero_maior_200 = filmes[filmes.duration >= 200]['genre'] #Usando []
```

```python
genero_maior_200
```

```python
#Porém, por boa prática, utilizamos sempre o método LOC que nada mais é do que definir uma condição e filtrar por labels
#Veremos o ILOC em mais detalhes

filmes.loc[filmes.duration >= 200, 'genre']
```
