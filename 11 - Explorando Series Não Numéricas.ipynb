# Explorando series nÃ£o numÃ©ricas

```python
# Comando 'describe' 
# (1) Conta todos os valores nÃ£o nulos de uma coluna. 
# (2) Mostra quais sÃ£o os valores Ãºnicos.
# (3) Qual dos valores se repete mais.
# (4) Frequencia do valor que se repete mais 

filmes.genre.describe()
```

```python
#Vamos verificar a frequencia dos valores
# VÃª a repetiÃ§Ã£o dos valores e a quantidade de cada um.

filmes.genre.value_counts()
```

```python
#Podemos exibir a porcentagem
# VÃª a portcentagem de repetiÃ§Ã£o dos valores e a quantidade de cada um.

filmes.genre.value_counts(normalize=True)
```

# Como podemos ver, value_counts() retorna uma sÃ©rie, e tal pode ser aproveitada com todos os mÃ©todos

```python
type(filmes.genre.value_counts())
```

```python
#Um exemplo simples Ã© o head()

filmes.genre.value_counts().head(10)
```

```python
#Ou o tail()

filmes.genre.value_counts().tail(10)
```

# unique() e nunique()

```python
#O mÃ©todo unique mostra os valores da sÃ©rie sem repetir

filmes.genre.unique()
```
