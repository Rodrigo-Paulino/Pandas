# Filtros MÃºltiplos

```python
import numpy as np
import pandas as pd
```

```python
colunas_filmes = ['star_rating','title','content_rating','genre','duration','actors_list']

file="C:\\. ProjetosDataScience\\Datasets\\movies.csv"
filmes = pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns 
```

```python
#Filtrando com duraÃ§Ã£o superior a 180 minutos

filmes[filmes.duration >= 180]
```

Entendendo **operadores lÃ³gicos:**

- **`and`**: Verdadeiro somente se **ambas as opÃ§Ãµes** forem verdadeiras
- **`or`**: Verdadeiro se **ao menos** uma das opÃ§Ãµes forem verdadeiras

```python
#Operador AND

print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
#Operador OR

print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

Regras de utilizaÃ§Ã£o de **mÃºltiplos filtros** no pandas:

- use **`&`** em vez de **`and`**
- use **`|`** em vez de **`or`**
- adicione **parenteses** em volta de cada critÃ©rio especÃ­fico, por ordem

### Trazer somente os filmes do gÃªnro drama E com duraÃ§Ã£o superior a 200 minutos

```python
dramas_longos = filmes[(filmes.duration >= 200) & (filmes.genre == 'Drama')]
```

```python
dramas_longos
```

```python
#Trocando a condiÃ§Ã£o AND por OR

dramas_longos = filmes[(filmes.duration >=200) | (filmes.genre == 'Drama')]
dramas_longos
```

### Trazer somente os filmes do gÃªnro drama OU com duraÃ§Ã£o superior a 200 minutos

```python
filmes_longos = filmes[(filmes.duration >=200) | (filmes.genre == 'Drama')]
```

```python
filmes_longos.head(10)
```

### Trazer os filmes dos gÃªneros Crime, Drama ou Action

```python
filmes[(filmes.genre == 'Crime') | (filmes.genre == 'Drama') | (filmes.genre == 'Action')].head(10)
```

### Utilizando o mÃ©todo ISIN

```python
# MÃ©todo ISIN = EstÃ¡ dentro

filmes[filmes.star_rating.isin([9.0,9.1,9.2,9.3,9.4,9.5])]
```

```python
filmes[filmes.genre.isin(['Crime', 'Drama', 'Action'])].head(10)
```
